```python
import os
import zipfile
import pandas as pd

# First, let's see what files are currently in your directory.
# This helps you identify the exact name of your uploaded zip file.
print("Files in current directory:")
print(os.listdir())

# --- Replace 'your_zip_file.zip' with the actual name of your zip file ---
zip_file_name = 'Retail Inventory Data (1).zip' # Example zip file name
# If you have another zip file, update this variable accordingly

# Create a directory to extract files into (optional but good practice)
extraction_path = 'extracted_data'
os.makedirs(extraction_path, exist_ok=True)

try:
    with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
        zip_ref.extractall(extraction_path)
    print(f"\nSuccessfully extracted '{zip_file_name}' to '{extraction_path}'.")

    # List files inside the extracted directory to find your data file
    print(f"\nFiles in '{extraction_path}' after extraction:")
    print(os.listdir(extraction_path))

    # --- Assuming your data is in a CSV file named 'your_data.csv' inside the zip ---
    # --- You'll need to replace 'your_data.csv' with the actual name of your data file ---
    data_file_in_zip = 'file_name_inside_zip.csv' # Example data file name
    full_file_path = os.path.join(extraction_path, data_file_in_zip)

    if os.path.exists(full_file_path):
        # Check if it's a CSV or Excel and use the appropriate pandas function
        if data_file_in_zip.endswith('.csv'):
            df_from_zip = pd.read_csv(full_file_path)
            print(f"\nSuccessfully loaded '{data_file_in_zip}' into a DataFrame.")
            display(df_from_zip.head())
        elif data_file_in_zip.endswith('.xlsx') or data_file_in_zip.endswith('.xls'):
            df_from_zip = pd.read_excel(full_file_path)
            print(f"\nSuccessfully loaded '{data_file_in_zip}' into a DataFrame.")
            display(df_from_zip.head())
        else:
            print(f"Unsupported file type for '{data_file_in_zip}'. Please load manually.")
    else:
        print(f"\nError: Data file '{data_file_in_zip}' not found in '{extraction_path}'. Please check the filename inside your zip.")

except FileNotFoundError:
    print(f"\nError: The zip file '{zip_file_name}' was not found. Please ensure it's uploaded or the name is correct.")
except Exception as e:
    print(f"\nAn unexpected error occurred: {e}")
```
